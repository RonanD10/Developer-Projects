```python
with open('/content/drive/MyDrive/Colab Notebooks/Scrabble/J-list.txt') as f:
    J_list = f.read().splitlines()

with open('/content/drive/MyDrive/Colab Notebooks/Scrabble/Q-list.txt') as f:
    Q_list = f.read().splitlines()

with open('/content/drive/MyDrive/Colab Notebooks/Scrabble/X-list.txt') as f:
    X_list = f.read().splitlines()

with open('/content/drive/MyDrive/Colab Notebooks/Scrabble/Z-list.txt') as f:
    Z_list = f.read().splitlines()
```

```python
J_words = [i.split('\t', 1)[0] for i in J_list]
J_play =  [i.split('\t', 1)[1] for i in J_list]
J_dict = dict(zip(J_words, J_play))

Q_words = [i.split('\t', 1)[0] for i in Q_list]
Q_play =  [i.split('\t', 1)[1] for i in Q_list]
Q_dict = dict(zip(Q_words, Q_play))

X_words = [i.split('\t', 1)[0] for i in X_list]
X_play =  [i.split('\t', 1)[1] for i in X_list]
X_dict = dict(zip(X_words, X_play))

Z_words = [i.split('\t', 1)[0] for i in Z_list]
Z_play =  [i.split('\t', 1)[1] for i in Z_list]
Z_dict = dict(zip(Z_words, Z_play))
```

```python
# Delete redundant suffexes

for word in J_words:
    if word.endswith('S') or word.endswith('ER') or word.endswith('ING') or word.endswith('ED'):
        J_words.remove(word)

for word in Q_words:
    if word.endswith('S') or word.endswith('ER') or word.endswith('ING') or word.endswith('ED'):
        Q_words.remove(word)

for word in X_words:
    if word.endswith('S') or word.endswith('ER') or word.endswith('ING') or word.endswith('ED'):
        X_words.remove(word)

for word in Z_words:
    if word.endswith('S') or word.endswith('ER') or word.endswith('ING') or word.endswith('ED'):
        Z_words.remove(word)
```

```python
def boring_affix(pref, suff):
    common_affix = ['S', 'ED', 'ING', 'Y', 'UN', 'IN', 'IER', 'ISH', 'EST', 'LY',
                    'ER', 'ES', 'ABLE', 'RE', 'ABLY', 'D', 'IVE', 'TION', 'AL',
                    'ISE', 'IZE', 'IST', 'ISM', 'MENT', 'ATE', 'AN', 'OR', 'ATE', 'ION']

    for aff in common_affix:
        if pref.startswith(aff) or suff.endswith(aff):
            return True

    return False
```

```python
for j_sub in J_words:
    if len(j_sub) > 8:
        break

    for j in J_words:
        if j_sub in j:
            affixes = j.split(j_sub)
            affixes.insert(1, j_sub)

            if len(j) < 8:
                continue

            if boring_affix(affixes[0], affixes[2]):
                continue

            if len(affixes[0]) >= 1 and len(affixes[2]) >= 1:
                 print(affixes[0] + '-' + affixes[1] + '-' + affixes[2])

            elif len(affixes[0]) == 0 and len(affixes[2]) >= 1:
                 print(affixes[1] + '-' + affixes[2])

            elif len(affixes[0]) >= 1 and len(affixes[2]) == 0:
                 print(affixes[0] + '-' + affixes[1])

            else:
                continue
```
